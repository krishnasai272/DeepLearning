```python
#DECISION TREE
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
data = {
    'Time': [1, 2, 3, 4, 5, 6],
    'Battery Charge': [12, 24, 36, 48, 60, 71]
}
df = pd.DataFrame(data)

X = df[['Time']]
Y = df['Battery Charge']
model = DecisionTreeRegressor()
model.fit(X, Y)
prediction_input = pd.DataFrame({'Time': [1.5]})
prediction = model.predict(prediction_input)
print (f"{prediction[0]}")
```
