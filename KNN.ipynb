```python
#KNN
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
data = {
    'Time': [1,3,5,7,4,6],
    'Battery Charge': [12, 24, 36, 48, 60, 71]
}
df = pd.DataFrame(data)

X = df[['Time']]
Y = df['Battery Charge']
model = KNeighborsRegressor(n_neighbors=2)
model.fit(X, Y)
prediction = model.predict([[2]])
print(f"Predicted Battery Charge at Time 7: {prediction[0]}")
```
